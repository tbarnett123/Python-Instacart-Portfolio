Number 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
ords_prods_merged_NV['avg_product_price'] = ords_prods_merged_NV.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merged_NV [['user_id','avg_product_price']].head(20)
```

```python
ords_prods_merged_NV [['user_id','avg_product_price']].tail(20)
```

```python
# Creating a 'Spending' Label Flag using loc Function
ords_prods_merged_NV.loc[ords_prods_merged_NV['avg_product_price'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merged_NV.loc[ords_prods_merged_NV['avg_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#checking the value count on the spending flag column
ords_prods_merged_NV['spending_flag'].value_counts(dropna = False)
```

Number 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

-If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”

-If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”

-If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
ords_prods_merged_NV['median_days_between_orders'] = ords_prods_merged_NV.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
#checking output
ords_prods_merged_NV[['user_id', 'days_since_prior_order','median_days_between_orders']].head(17)
```

```python
ords_prods_merged_NV[['user_id', 'days_since_prior_order','median_days_between_orders']].tail(17)
```

```python
#creating an order frequncy label flag
ords_prods_merged_NV.loc[ords_prods_merged_NV['median_days_between_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merged_NV.loc[(ords_prods_merged_NV['median_days_between_orders'] <= 20) & (ords_prods_merged_NV['median_days_between_orders'] > 10), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merged_NV.loc[ords_prods_merged_NV['median_days_between_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#checking output of value counts on the order frequency label flag
ords_prods_merged_NV ['order_frequency_flag'].value_counts(dropna = False)
```

```python
#checking dataframes created to ensure they have updated
ords_prods_merged_NV [['user_id', 'days_since_prior_order','median_days_between_orders','order_frequency_flag',]].head(20)
```
