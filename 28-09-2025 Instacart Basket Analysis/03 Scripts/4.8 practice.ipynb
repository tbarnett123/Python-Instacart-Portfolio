```python
orders_products_merge_NV.loc[orders_products_merge_NV['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
orders_products_merge_NV.loc[(orders_products_merge_NV['max_order'] <= 40) & (orders_products_merge_NV['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
orders_products_merge_NV.loc[orders_products_merge_NV['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
