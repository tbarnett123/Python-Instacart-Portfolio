```python
#Randomized number that represents a sample in order to be able to accurately make a line chart. This is the code to be able to do that.
np.random.seed(4)
dev = np.random.rand(len(ords_prods_merged_NVG)) <= 0.7
```

the np.random.rand() produces a list of random numbers between 1 and 0. Here the agrument is 10, so 10 random numbers are produced. In the code above, the argument given is the length of your dataframe.

So, to recap each piece of the split code, you need to:

1.Set the seed for the random number generator.

2.Create 32,404,859 random numbers (the length of your ords_prods_merge dataframe) ranging between 0 and 1.

3.Code a logical test comparing each of those numbers to a value of 0.7 (either less than or greater than 0.7).

4.Store the results of the test in a new list, dev, in the form of boolean values (True and False).

To use the list to split the two dataframes into two samples

big = ords_prods_merged_NVG[dev]
small = ords_prods_merged_NVG[~dev]

store 70 percent of your dataframe in the big sample and the 30 percent in the small dataframe
for every value in the list that is true, the row in ords_prods_merged_NVG will be stored in the big sample. For every value thats False, the corresponding row in the dame dataframe will be stored in the small sample.

The tilde stores everything thats NOT dev, or false.

```python
#create a list holding true/false values to the test np.random.rant() <=0.7
dev = np.random.rand(len(ords_prods_merged_NVG)) <=0.7
```

```python
#Store 70% of the sample in the dataframe big
big = ords_prods_merged_NVG[dev]
```

```python
#Store 30% of the sample in the dataframe small
small = ords_prods_merged_NVG[~dev]
```

```python
len(ords_prods_merged_NVG)
```

```python
len(big) + len(small)
```

```python
#this step will reduce the samples to only those columns necessary for the chart, namely prices and orders day of week.
df_2 = small[['orders_day_of_week','prices']]
```

```python
#the dataframe above only contains the prices and orders day of week columns from the small subset
```

```python
#creating a line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```

sns.lineplot() function is being used to create the line chart and save it to a new object,line. The arguments it needs are the dataframe you want to create the line chart from (df_2), the column you want to place on the x-axis (orders_day_of_week) and the column you want to placve on the y-axis (prices).

You may be wondering what those translucent blue bands are around your line. Those are called error bands, and they represent the charts confidence intervals. In other words, if this line were a prediction, there would be a 95 percent probability that the predicted value would fall within the area of the error bands.
